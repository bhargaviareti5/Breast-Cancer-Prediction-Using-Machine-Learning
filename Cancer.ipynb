```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Store object column names
original_columns = df.select_dtypes(include='object').columns

# Initialize LabelEncoder
label_encoders = {}

# Apply LabelEncoder to each categorical variable
for col in original_columns:
    # Replace NaN with a placeholder and convert to string
    df[col] = df[col].fillna('missing').astype(str)
    
    # Initialize and apply LabelEncoder
    label_encoders[col] = LabelEncoder()
    df[col] = label_encoders[col].fit_transform(df[col])

# Print the mapping between original categories and numerical labels
for col, encoder in label_encoders.items():
    print(f"Mapping for column '{col}':")
    for label, category in enumerate(encoder.classes_):
        print(f"Label {label}: {category}")
    print("=====================")
```

# *4. Data Splitting*

```python
x = df.drop('diagnosis', axis=1)
y = df['diagnosis']
```

```python
y.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(random_state=42)
x,y = smote.fit_resample(x,y)
```

```python
y.value_counts()
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
```

```python
selected = SelectKBest(score_func=f_classif, k = 15)
x_selected = selected.fit_transform(x,y)
```

```python
selected_index = selected.get_support(indices=True)
print(selected_index)
```

```python
selected_columns = x.columns[selected.get_support()] 
print(selected_columns.to_list())
```

```python
df_selected = pd.DataFrame(x_selected, columns=selected_columns)
df_selected['Class'] = y.values
```

```python
df_selected.to_csv('Final_cancer.csv', index=False)
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_selected,y, test_size=0.2, random_state=42)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

# Hybrid Algorithm (CNN + Random Forest)

```python
# Standardize the features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(x_train)
X_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense
from tensorflow.keras import optimizers

# Define the CNN model for tabular data
cnn_model = Sequential()
cnn_model.add(Conv1D(64, 3, activation='relu', input_shape=(X_train.shape[1], 1)))  # 1D convolution
cnn_model.add(MaxPooling1D(2))
cnn_model.add(Flatten())
cnn_model.add(Dense(64, activation='relu'))
cnn_model.add(Dense(32, activation='relu'))
cnn_model.add(Dense(1, activation='sigmoid'))  # For binary classification; 

# Compile the CNN model
cnn_model.compile(optimizer=optimizers.Adam(), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Reshape data for CNN input (samples, features, 1 channel)
X_train_cnn = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test_cnn = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))
```

```python
cnn_model.fit(X_train_cnn, y_train, epochs=10, batch_size=32, validation_data=(X_test_cnn, y_test))
```

```python
# Extract features from CNN
cnn_features_train = cnn_model.predict(X_train_cnn)
cnn_features_test = cnn_model.predict(X_test_cnn)

# Flatten the CNN features to 1D (as required for Random Forest)
cnn_features_train = cnn_features_train.flatten().reshape(-1, 1)
cnn_features_test = cnn_features_test.flatten().reshape(-1, 1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Initialize and train the Random Forest classifier
rf = RandomForestClassifier()
rf.fit(cnn_features_train, y_train)

# Predict on the training data
train_pred = rf.predict(cnn_features_train)
train_accuracy = accuracy_score(y_train, train_pred)
print(f"Training accuracy of Hybrid Model (CNN + RF): {train_accuracy}")

# Classification report
train_report = classification_report(y_train, train_pred)
print("Classification Report (Training):\n", train_report)

# Confusion matrix for training data
train_conf_matrix = confusion_matrix(y_train, train_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(train_conf_matrix, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix (Training)")
plt.show()

# Predict on the test data
test_pred = rf.predict(cnn_features_test)
test_accuracy = accuracy_score(y_test, test_pred)
print(f"Testing accuracy of Hybrid Model (CNN + RF): {test_accuracy}")

# Classification report
test_report = classification_report(y_test, test_pred)
print("Classification Report (Testing):\n", test_report)

# Confusion matrix for testing data
test_conf_matrix = confusion_matrix(y_test, test_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(test_conf_matrix, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix (Testing)")
plt.show()
```
